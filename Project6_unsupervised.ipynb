# 2. Encode independent features.

```python
df1 = pd.get_dummies(data = df)
```

```python
df1.head()
```

```python
df1.shape
```

```python
df1.describe()
```

# 3. Separate the dependant variable from the rest of the train 

```python
from sklearn.model_selection import train_test_split
x = np.array(df1)
y = np.array(dataset['NB/B'])
```

# 4. Split the dataset into training and test data (Take 80:20 ratio)

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.20, random_state = 1)  
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

#  5. Apply PCA on the independent features of train data.

```python
from sklearn.decomposition import PCA
```

```python
P = PCA().fit(X_train)
```

#  6. Display elbow plot to find out reduced number of dimensions (which covers more than 95% of the variance). 

```python
plt.figure(figsize=(10,6))
plt.plot(np.cumsum(P.explained_variance_ratio_))

plt.xlabel('Number of components')
plt.ylabel('Total explained variance')

plt.xlim(0, 20)
plt.yticks(np.arange(0, 1.1, 0.2))

plt.axvline(10, c='g')
plt.axhline(0.96, c='r')
```

```python
#the number of components which cover more than 95% is near around 10
# So there are 10 Principle components.
```

 # 7. Transform the data using PCA (select optimum number of PCs). 

```python
PC= PCA(n_components=10)
```

```python
PC.fit(X_train)
```

```python
XP= PC.transform(X_train)
```

```python
XP.shape
```

```python
XP_d=pd.DataFrame(XP, columns=['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9','PC10'])
```

```python
XP_d.head()
```

```python
sns.pairplot(XP_d,diag_kind='kde')
```

```python
PC.explained_variance_ratio_
```

 # 8. Apply KMeans algorithm on the above transformed data. 

```python
from sklearn.cluster import KMeans

Cluster_range = range( 1, 15)
Cluster_errors = []
cluster_sil_scores = []
for num_clusters in Cluster_range:
  clusters = KMeans( num_clusters, n_init = 5)
  clusters.fit(XP_d)
  labels = clusters.labels_                    
  centroids = clusters.cluster_centers_ 
  Cluster_errors.append( clusters.inertia_ )
  
Clusters_df = pd.DataFrame( { "Num_clusters":Cluster_range, "Cluster_errors": Cluster_errors} )
Clusters_df[0:15]
```

```python
plt.figure(figsize=(10,6))
plt.plot( Clusters_df.Num_clusters, Clusters_df.Cluster_errors, marker = "o" )
```

```python
#maximum bend at 2
#hence selecting 2 cluster
```

#  9. Select optimal number of clusters and form clusters out of the train data

```python
K_mean = KMeans(n_clusters=2, n_init = 15, random_state=1)
K_mean.fit(XP_d)
```

```python
print(K_mean.cluster_centers_)
```

# 10. Create a new column in the train data and populate it with the cluster labels. 

```python
km_df= pd.DataFrame(K_mean.labels_ , columns = list(['labels']))
km_df.head()
```

```python
XP_d["Clus_Labels"]=km_df['labels']
```

 # 13. Apply SVC model on train dataset and predict target for test dataset. 


```python
from sklearn.svm import SVC
S = SVC()
S.fit(XP_d, Y_train)
Y_predict = S.predict(XP_test_DF)
print(S.score(XP_d, Y_train))
print(S.score(XP_test_DF, Y_test))
```
